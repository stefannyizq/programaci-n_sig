```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point
from shapely import affinity
import contextily as ctx

plt.rcParams['figure.figsize'] = (10, 8)
```

```python
#1. Preparación de Datos
# Define rutas a los archivos

SHAPE_MUN = 'datos_colombia/municipios_colombia.shp'
EVA_CSV = 'datos_colombia/Evaluaciones_Agropecuarias_Municipales_EVA.csv'
EST_CSV = 'datos_colombia/Catalogo_Estaciones_IDEAM.csv'
ROADS_SHP = 'datos_colombia/RedVial.shp'

# Municipios
gdf_mun = gpd.read_file(SHAPE_MUN).to_crs(epsg=9377)

# EVA
eva = pd.read_csv(EVA_CSV)
eva.rename(columns={"CÓD. MUN.": "cod_mun"}, inplace=True)
eva['cod_mun'] = eva['cod_mun'].astype(str).str.zfill(5)

# Estaciones IDEAM
df_est = pd.read_csv(EST_CSV)
coords = df_est['Ubicación'].str.strip('()').str.split(',', expand=True).astype(float)
df_est['lat'], df_est['lon'] = coords[0], coords[1]
gdf_est = gpd.GeoDataFrame(df_est, geometry=gpd.points_from_xy(df_est['lon'], df_est['lat']), crs="EPSG:4326")
gdf_est = gdf_est.to_crs(epsg=9377)

print(f"{len(gdf_mun):,} municipalities • {len(gdf_est):,} stations • "
      f"{len(eva):,} EVA rows • {len(roads):,} vias")

# Carreteras de INVIAS
roads = gpd.read_file(ROADS_SHP).to_crs(epsg=9377)

#Homogenise municipalities codes
gdf_mun["MPIO_CCDGO"] = gdf_mun["MPIO_CCDGO"].astype(str).str.zfill(5)
eva["cod_mun"]        = eva["cod_mun"].astype(str).str.zfill(5)
```

```python
#Criterios de Análisis a nivel municpal
```

```python
# 1. Spatial join: each station takes the attributes of the municipality it falls in
stations_with_mun = gpd.sjoin(
    gdf_est,                     # left: point layer
    gdf_mun[["MPIO_CCDGO", "MPIO_CNMBR","MPIO_CCDGO", "geometry"]],  # right: polygon layer
    how="left",
    predicate="within"           # point inside polygon
)
est_counts = stations_per_mun.groupby("MPIO_CCDGO").size().reset_index(name='ideam_count')

# 2. Inspect the result
stations_with_mun[["Codigo", "MPIO_CNMBR", "MPIO_CCDGO"]].head()
```

```python
# Aseguramos que los códigos sean strings
gdf_mun['MPIO_CCDGO'] = gdf_mun['MPIO_CCDGO'].astype(str).str.zfill(5)
gdf_mun['DPTO_CCDGO'] = gdf_mun['DPTO_CCDGO'].astype(str).str.zfill(2)

# Filtramos Valle del Cauca (DPTO_CCDGO = '76')
valle_mun = gdf_mun[gdf_mun['DPTO_CCDGO'] == '76']

# Visualizamos para confirmar
ax = valle_mun.plot(color='lightblue', edgecolor='black')
plt.title("Municipios de Valle del Cauca")
plt.show()

# Imprimimos cuántos municipios tenemos
print(f"Total de municipios en Valle del Cauca: {len(valle_mun)}")
```

```python
# Caña de azúcar
eva_cana = eva[eva["CULTIVO"].str.contains("CAÑA", case=False)]
eva_cana_sum = eva_cana.groupby("cod_mun")['Producción\n(t)'].sum().reset_index()


for col in ['cod_mun', 'Producción\n(t)']:
    if col in valle_mun.columns:
        valle_mun = valle_mun.drop(columns=[col])

valle_mun = valle_mun.merge(eva_cana_sum, left_on="MPIO_CCDGO", right_on="cod_mun", how="left")
valle_mun['Producción\n(t)'] = valle_mun['Producción\n(t)'].fillna(0)
```

```python
# Limpieza previa antes del merge de estaciones
if 'num_stations' in valle_mun.columns:
    valle_mun = valle_mun.drop(columns=['num_stations'])

# Spatial join
stations_with_mun = gpd.sjoin(
    gdf_est, 
    valle_mun[['MPIO_CCDGO', 'geometry']], 
    how="left", 
    predicate="within"
)

station_counts = stations_with_mun.groupby("MPIO_CCDGO").size().reset_index(name="num_stations")
valle_mun = valle_mun.merge(station_counts, on="MPIO_CCDGO", how="left")
valle_mun['num_stations'] = valle_mun['num_stations'].fillna(0)
```

```python
roads_in_valle = gpd.overlay(roads, valle_mun, how='intersection')
roads_in_valle['length_m'] = roads_in_valle.geometry.length

road_density = roads_in_valle.groupby("MPIO_CCDGO")['length_m'].sum().reset_index()
valle_mun = valle_mun.merge(road_density, on="MPIO_CCDGO", how="left")
valle_mun['length_m'] = valle_mun['length_m'].fillna(0)

valle_mun['area_km2'] = valle_mun.area / 1e6
valle_mun['road_density'] = valle_mun['length_m'] / valle_mun['area_km2']
valle_mun['road_density'] = valle_mun['road_density'].fillna(0)
```

```python
# Normalización robusta
max_cana = valle_mun['Producción\n(t)'].max()
valle_mun['norm_cana'] = valle_mun['Producción\n(t)'] / max_cana if max_cana > 0 else 0

max_stations = valle_mun['num_stations'].max()
valle_mun['norm_stations'] = valle_mun['num_stations'] / max_stations if max_stations > 0 else 0

max_roads = valle_mun['road_density'].max()
valle_mun['norm_roads'] = valle_mun['road_density'] / max_roads if max_roads > 0 else 0


valle_mun['ranking'] = (w_cana * valle_mun['norm_cana'] +
                        w_stations * valle_mun['norm_stations'] +
                        w_roads * valle_mun['norm_roads'])

# Selección del municipio óptimo
best_mun = valle_mun.loc[valle_mun['ranking'].idxmax()]
print("✅ Municipio óptimo seleccionado:")
print(best_mun[['MPIO_CNMBR', 'ranking']])
```

```python
#Se seleccion apalmira por que tiene alta productividad agropecuaria de caña de azúcar,buena infraestructura vial,Y disponibilidad de datos climáticos.
```

```python
#Level 2

from shapely.geometry import box

# Primero extraemos la geometría de Palmira
best_mun_gdf = valle_mun[valle_mun['MPIO_CNMBR'] == 'PALMIRA'].copy()

# Definimos resolución del grid (en metros)
grid_resolution = 1000  # cada 1 km

# Creamos el bounding box del municipio
minx, miny, maxx, maxy = best_mun_gdf.total_bounds

# Generamos lista de puntos de la malla
x_coords = np.arange(minx, maxx, grid_resolution)
y_coords = np.arange(miny, maxy, grid_resolution)

grid_points = []
for x in x_coords:
    for y in y_coords:
        point = Point(x, y)
        if best_mun_gdf.contains(point).any():
            grid_points.append(point)

# Convertimos a GeoDataFrame
grid_gdf = gpd.GeoDataFrame(geometry=grid_points, crs=best_mun_gdf.crs)

# Visualizamos
base = best_mun_gdf.plot(color='lightgreen', edgecolor='black')
grid_gdf.plot(ax=base, color='red', markersize=5)
plt.title("Grid de análisis en Palmira")
plt.show()
```

```python
from shapely.ops import nearest_points

# Filtramos las vías dentro de Palmira
roads_in_palmira = gpd.overlay(roads, best_mun_gdf, how='intersection')

# Filtramos estaciones dentro de Palmira
stations_in_palmira = gpd.sjoin(gdf_est, best_mun_gdf, how='inner', predicate='within')

# Calculamos distancia a vías
grid_gdf['dist_road'] = grid_gdf.geometry.apply(lambda point: roads_in_palmira.distance(point).min())

# Calculamos distancia a estaciones
grid_gdf['dist_station'] = grid_gdf.geometry.apply(lambda point: stations_in_palmira.distance(point).min())

# Visualización simple de distancias a carreteras
grid_gdf.plot(column='dist_road', cmap='viridis', legend=True, markersize=5)
plt.title("Distancia a carreteras")
plt.show()

# Visualización simple de distancias a estaciones
grid_gdf.plot(column='dist_station', cmap='plasma', legend=True, markersize=5)
plt.title("Distancia a estaciones IDEAM")
plt.show()
```

```python
# Normalización inversa (entre 0 y 1, donde 1 = más cerca)
grid_gdf['norm_road'] = 1 - (grid_gdf['dist_road'] / grid_gdf['dist_road'].max())
grid_gdf['norm_station'] = 1 - (grid_gdf['dist_station'] / grid_gdf['dist_station'].max())

# Ponderación de los criterios
w_road = 0.6
w_station = 0.4

grid_gdf['aptitude'] = (w_road * grid_gdf['norm_road'] +
                         w_station * grid_gdf['norm_station'])

# Selección del punto óptimo
best_point = grid_gdf.loc[grid_gdf['aptitude'].idxmax()]

print("Punto óptimo dentro de Palmira:")
print(best_point)

# Visualización del punto óptimo
base = best_mun_gdf.plot(color='lightgreen', edgecolor='black')
grid_gdf.plot(ax=base, column='aptitude', cmap='YlOrRd', markersize=10, legend=True)
gpd.GeoSeries(best_point.geometry).plot(ax=base, color='blue', markersize=50)
plt.title("Ubicación óptima dentro de Palmira")
plt.show()
```
